## 3. Definir la clase Neo4jConnection

```python
class Neo4jConnection:

    def __init__(self, uri, user, pwd):
        self.__uri = uri
        self.__user = user
        self.__pwd = pwd
        self.__driver = None

        try:
            self.__driver = GraphDatabase.driver(self.__uri, auth=(self.__user, self.__pwd))
        except Exception as e:
            print("Failed to create the driver:", e)

    def close(self):
        if self.__driver is not None:
            self.__driver.close()

    def query(self, query, parameters=None, db=None):
        assert self.__driver is not None, "Driver not initialized!"
        session = None
        response = None

        try:
            session = self.__driver.session(database=db) if db is not None else self.__driver.session()
            response = list(session.run(query, parameters))
        except Exception as e:
            print("Query failed:", e)
        finally:
            if session is not None:
                session.close()
        return response
```

## 4. Ejecutar operaciones de consulta

```python
# Crear instancia de la clase
conn = Neo4jConnection(uri=uri, user=username, pwd=password)

# Consulta para contar nodos y relaciones
query = ''' MATCH (n), ()-[r]->()
            RETURN count(n) AS numNodos, count(r) AS numRelaciones'''

result = conn.query(query)
print("Resultado de la consulta:")
for record in result:
    print(record)
```

## 5. Cerrar la conexión

```python
# Cerrar la conexión
conn.close()
driver.close()

print("Conexiones cerradas correctamente")
```
